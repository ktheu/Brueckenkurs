### Passwörter und Passkeys

#### Das Problem mit Passwörtern

- Wer verwendet überall dasselbe Passwort?
- Wer hat schon einmal ein Passwort vergessen?
- Was passiert, wenn eine Website gehackt wird? 


Mit "eine Website wurde gehackt" meint man hier: der Angreifer hat Zugriff auf die Datenbank der Webseite erhalten haben.

Darin können stehen:
- Benutzernamen
- E-Mail-Adressen
- Passwort-Hashes


Was kann passieren, wenn der Angreifer die Passwort-Hashes stiehlt?

Der Angreifer kennt das Passwort nicht direkt, kann aber versuchen, häufige Passwörter auszuprobieren und zu hashen, bis ein Treffer entsteht (Dictionary- oder Brute-Force-Angriff).

Schwache Passwörter werden dabei oft schnell geknackt.

#### Dictionary-Angriff

Menschen wählen oft vorhersehbare Passwörter:

    123456
    password
    hallo123
    sommer2025
    schalke04

Ein Angreifer erstellt deshalb ein „Wörterbuch“ mit Millionen häufiger Passwörter und testet diese automatisch.

```python
'''
Dictionary Angriff
'''
import hashlib

# Das echte Passwort des Benutzers
passwort = "hallo123"

# Die Webseite speichert nur den Hash
gespeicherter_hash = hashlib.sha256(passwort.encode()).hexdigest()

print("Gestohlener Hash:")
print(gespeicherter_hash)
print()

# Wörterbuch mit häufigen Passwörtern
woerterbuch = [
    "123456",
    "passwort",
    "hallo",
    "qwertz",
    "sommer2025",
    "hallo123",
    "geheim"
]

print("Starte Dictionary-Angriff...\n")

for kandidat in woerterbuch:
    kandidat_hash = hashlib.sha256(kandidat.encode()).hexdigest()

    print(f"Teste: {kandidat:20} berechneter Hash: {kandidat_hash}")

    if kandidat_hash == gespeicherter_hash:
        print()
        print("PASSWORT GEFUNDEN!")
        print("Passwort:", kandidat)
        break
```

Schutzmaßnahmen dagegen:

    - lange Passwörter:
        Blauer-Hund-rennt-2025-zum-See
    - zufällige Passwörter
        xT7#Lm9@Q2
    - Passwortmanager:
        Sie erzeugen lange zufällige Passwörter, die kaum durch Dictionary- oder Brute-force-Angriffe gefunden werden können.

##### Gängige Methoden für lange Passwörter

Satzmethode:
 
    Ich fahre jeden Morgen um 7 Uhr mit dem Fahrrad zur Schule!
    IfjMu7UmdFzS!

Gedicht-, Lied-, Buchtexte:

    Wie froh bin ich, daß ich weg bin!  (Die Leiden des jungen Werther, 1774)
    Wfbidiwb!1774

Mehrere zufällige Wörter kombinieren.

    Kaktus-Kanu-Lampe-Wolke




Wenn man mit langen, zufälligen Passwörtern arbeitet, bleiben trotzdem noch Probleme:

#### Phishing

Du erhältst eine E-Mail: "Ihr Konto wurde gesperrt. Bitte melden Sie sich an."
Der Link führt auf eine gefälschte Webseite. 

#### Malware

Wenn Schadsoftware auf dem Computer läuft, kann sie möglicherweise:
    
    - Tastatureingaben mitlesen,
    - Browserdaten auslesen,
    - Sitzungen übernehmen.

Ein starkes Passwort schützt dagegen nicht.

##### Exkurs: Was bedeutet es "eine Sitzung übernehmen"?



Browser sendet:

    Benutzername: max
    Passwort: geheim123

Server prüft Daten:

    Passwort korrekt

Beim nächsten Klick:

    GET /warenkorb

würde der Server Max bereits wieder vergessen haben. Das Internet-Protokoll HTTP ist zustandslos (stateless). Jede Anfrage sieht zunächst unabhängig aus. Man müsste theoretisch bei jeder Aktion erneut Benutzername und Passwort senden.
Das wäre unpraktisch und unsicher.

Lösung: Session-Cookies

Nach erfolgreichem Login erzeugt der Server eine zufällige Kennung, die Session-ID, und sendet diese an den Browser zurück. Der Browser speichert diese ID als Session-Cookie. In der weiteren Kommunikation sendet der Browser von Max die Session-ID an den Server, damit der Server weiß, dass die Anfrage von Max kommt. 

Wenn Malware diese Session-ID ausliest, kann der Angreifer eine Session im Namen von Max aufbauen, ohne ein Passwort eingeben zu müssen.

##### Exkurs: Cookies


Welche Arten von Cookies gibt es?
1. Session-Cookies
   - nur während der aktuellen Sitzung
   - verschwinden meist beim Schließen des Browsers
 
Beispiel:

    session=A7F83D91C2B4


2. Permanente Cookies
    - besitzen ein Ablaufdatum
    - bleiben Tage, Wochen oder Monate erhalten

Beispiel:

    sprache=de

Typische Anwendungen:

    session=A7F83D91C2B4    # Der Server erkennt den Benutzer wieder
    warenkorb=12345         # Die ausgewählten Produkte bleiben erhalten.
    sprache=de              # Die Webseite erscheint direkt auf Deutsch.

Ein Cookie enthält normalerweise nur Text. Es ist kein Programm und kann nicht selbstständig ausgeführt werden. Das Passwort dient dazu, die
Identität nachzuweisen, das Cookie dient dazu, den Benutzer zwischen Anfragen wiederzuerkennen.


Werbeanbieter können Cookies nutzen, um Nutzer über verschiedene Webseiten hinweg wiederzuerkennen.

##### Exkurs: Tracking Cookies

In Europa gilt die DSGVO sowie die ePrivacy-Richtlinie. 
Für technisch notwendige Cookies braucht man oft keine Zustimmung.

Beispiele:

    Login-Cookies
    Warenkorb-Cookies

Für Tracking-Cookies ist meist eine Zustimmung erforderlich. 
Deshalb erscheinen Hinweise wie:

„Wir verwenden Cookies für Analyse- und Marketingzwecke.“

Werbeanbieter möchten wissen:

- Welche Webseiten besucht jemand?
- Wofür interessiert sich die Person?
- Welche Werbung wurde bereits angezeigt?
- Auf welche Werbung wurde geklickt?

Dazu müssen sie einen Nutzer über mehrere Webseiten hinweg wiedererkennen.

Beispiel:

Angenommen, es gibt die Werbefirma AdNet.

Du besuchst:
 
    www.fahrrad-shop.de

Auf der Seite ist Werbung von AdNet eingebunden.
Der Browser lädt dabei Inhalte von AdNet.
AdNet speichert ein Cookie (Drittanbieter-Cookie):

    id=123456

Später besuchst du:

    www.sport-news.de

Auch dort ist Werbung von AdNet eingebunden. Der Browser sendet automatisch

    id=123456

an AdNet. AdNet erkennt: Das ist derselbe Browser wie zuvor. 
Nach vielen Webseitenbesuchen entsteht ein Profil:

    Nutzer 123456

    Interessen:
    - Fahrräder
    - Mountainbike
    - Outdoor-Sport
    - Camping
    

Besuchst du später eine andere Webseite, zeigt AdNet Werbung an:

    Mountainbike-Angebote

anstatt zufälliger Werbung.

In Chrome:

#### 2-Faktor-Authentifizierung (2FA)


Bei einer normalen Anmeldung genügt ein einziges Geheimnis.
Wenn ein Angreifer das Passwort kennt, kann er sich anmelden.

Egal ob das Passwort gestohlen wurde durch:

    Phishing
    Datenleck
    Malware
    Schulterblick

Der Angreifer besitzt dann alles, was er braucht.

Die Idee der 2-Faktor-Authentifizierung: 
Man verlangt zwei unabhängige Nachweise. 
Typischerweise: 
    
    Faktor 1: Wissen - Etwas, das du weißt: Passwort
    Faktor 2: Besitz - Etwas, das du besitzt: Smartphone, TAN-App, Hardware-Token 


    

Welche Probleme bleiben trotz 2FA?

- Viele Systeme senden den 2. Faktor per SMS: Diese Codes können abgefangen oder erschlichen werden (z.B durch Malware auf dem Handy)
- Phishing: 
Angreifer können heutzutage oft beides abfragen: 
Passwort + aktuellen 2FA-Code auf einer auf einer gefälschten Webseite.
- SIM-Swapping: Hierbei versucht ein Angreifer, den Mobilfunkanbieter davon zu überzeugen, eine neue SIM-Karte für die Telefonnummer des Opfers auszustellen. Wenn dies gelingt: die neue SIM-Karte des Angreifers hat die Telefonnummer des Opfers

#### Passkeys

Ein Passkey besteht aus einem Schlüsselpaar:

- öffentlicher Schlüssel
- privater Schlüssel

Der private Schlüssel liegt auf dem Gerät des Benutzers: Smartphone, Tablet, Laptop

Der öffentliche Schlüssel wird beim Dienst gespeichert.

##### Beispiel:

Anmeldung bei einem Online-Shop.

Registrierung: Das Smartphone erzeugt ein Schlüsselpaar. Der öffentliche Schlüssel wird an den Shop gesendet.
Der private Schlüssel bleibt auf dem Smartphone.



Nun möchte sich der Benutzer anmelden:

- Der Shop erzeugt eine Zufallszahl (Challenge), z.B. 83746291
- Das Smartphone signiert diese Zahl mit dem privaten Schlüssel.
- Die Signatur wird an den Shop gesendet.
- Der Shop prüft die Signatur mit dem öffentlichen Schlüssel.
- Wenn die Prüfung erfolgreich ist: Dieses Gerät besitzt den passenden privaten Schlüssel, Anmeldung erfolgreich.

Passkeys verwenden die Idee der asymmetrischen Kryptografie, aber zum Nachweis der Identität statt zur Verschlüsselung.



##### Warum sind Passkeys sicherer?

- Kein Passwort kann gestohlen werden. 
Bei einer Phishing-Seite gibt es kein Passwort zum Eingeben.
Der Angreifer erhält nichts.
- Datenbank-Leaks sind weniger gefährlich. Der Server speichert nur öffentliche Schlüssel. 
Wer die Datenbank stiehlt, erhält keine privaten Schlüssel.
- Einzigartiger Schlüssel pro Website. 
Für jede Website wird ein eigenes Schlüsselpaar erzeugt.
Wenn ein Dienst kompromittiert wird, sind andere Konten nicht betroffen.

##### Typischer Ablauf auf dem Smartphone
- Website öffnen, „Mit Passkey anmelden“, Fingerabdruck oder Face-ID
- Smartphone verwendet den privaten Schlüssel, Anmeldung erfolgreich
- Der Fingerabdruck wird dabei nicht an die Website geschickt. 
Er dient nur dazu, den privaten Schlüssel auf dem Gerät freizugeben.

##### Wo werden Passkeys gespeichert?

Der private Schlüssel ist Teil einer geschützten Sicherheitsinfrastruktur des Geräts und wird normalerweise nie direkt sichtbar. Das ist einer der wesentlichen Unterschiede zu Passwörtern, die ein Benutzer kennen, aufschreiben oder versehentlich weitergeben kann.

Beispiele:

In Windows ist TPM (Trusted Platform Module) ein spezieller Sicherheitschip auf dem Mainboard (oder in moderneren CPUs/Firmwares integriert), der kryptographische Schlüssel besonders schützt speichert und Sicherheitsfunktionen bereitstellt.

Smartphones besitzen häufig spezielle Sicherheitschip zur Speicherung von Passkeys, die Apps nicht einfach auslesen können.



##### Synchronization

Passkeys können durch die Passwort-Manager des Betriebssystems synchronisiert werden. Die Synchronisierung erfolgt verschlüsselt über den jeweiligen Anbieter (Google, Apple, Microsoft), sodass die Schlüssel auf mehreren Geräten verfügbar werden, ohne dass sie offen einsehbar sind.

Es ist aber auch möglich, auf jedem Gerät einen eigenen Passkey für einen Dienst zu erstellen (mehrere Haustürschlüssel).
